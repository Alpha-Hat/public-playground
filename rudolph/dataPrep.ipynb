```python
import requests
from bs4 import BeautifulSoup


# Parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Extract story titles and URLs
stories_list = []
for link in soup.find_all("a"):
    title = link.get_text(strip=True)
    url = link["href"]
    stories_list.append({"title": title, "url": url})

# Base URL of the website
base_url = "https://americanliterature.com"  # Replace with the actual website's base URL

# Dictionary to store the stories
stories_dict = {}


def fetch_story_text(url):
    """
    Fetch the story text from the given URL.

    Args:
        url (str): The URL of the story.

    Returns:
        str: The extracted story text, or None if an error occurs.
    """
    
    try:
        # Send a GET request to the URL
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }
        response = requests.get(base_url + url, headers=headers, timeout=10)
        response.raise_for_status()  # Raise an error for HTTP status codes >= 400

        # Parse the HTML content
        soup = BeautifulSoup(response.text, "html.parser")

        # Extract text from all <p> tags in the story content
        paragraphs = soup.find_all("p")
        story_text = "\n".join(p.get_text(strip=True) for p in paragraphs)

        return story_text
    except requests.exceptions.HTTPError as http_err:
        print(f"Failed to fetch story from {url}: {http_err}")
        return None
    except requests.exceptions.RequestException as req_err:
        print(f"Error occurred while fetching {url}: {req_err}")
        return None
#url = "https://americanliterature.com/author/aesop/short-story/the-boy-and-the-filberts"

# Iterate through the list of stories
for story in stories_list:
    title = story["title"]
    url = story["url"]

    #print(f"Fetching story: {title} from {url}...")

    # Fetch the story content
    story_text = fetch_story_text(url)
    #print(base_url+url)

    # If the story was successfully fetched, add it to the dictionary
    if story_text:
        stories_dict[title] = story_text
        print("Story text extracted successfully:")
        #print(story_text)
        #for title, text in stories_dict.items():
            #print(f"Title: {title}\nText: {text[:100]}...")  # Print the first 100 characters of the story text

    else:
        print("Failed to extract story text.")

# Output the result
#   print("Stories imported into the dictionary:")
# for title, text in stories_dict.items():
#   print(f"Title: {title}\nText: {text[:100]}...")  # Print the first 100 characters of the story text


# Example usage
#story_url = "https://americanliterature.com/author/aesop/short-story/the-boy-and-the-filberts"
#story_text = fetch_story_text(story_url)

#if story_text:
#    print("Story text extracted successfully:")
#   print(story_text)
#else:
#    print("Failed to extract story text.")
```
